```python
import Bio
from Bio import SeqIO

# File paths for the two FASTA files
file1 = "test_pos.txt"
file2 = "test_neg.txt"

# Initialize a variable to store the maximum length
max_length = 0

# Iterate through the sequences in the first FASTA file
for record in SeqIO.parse(file1, "fasta"):
    # Update the maximum length if the current sequence is longer
    max_length = max(max_length, len(record.seq))

# Iterate through the sequences in the second FASTA file
for record in SeqIO.parse(file2, "fasta"):
    # Update the maximum length if the current sequence is longer
    max_length = max(max_length, len(record.seq))

# Print the maximum length
print("Maximum length:", max_length)
```

```python
import matplotlib.pyplot as plt
from Bio import SeqIO

# Parse FASTA file 1
sequences1 = SeqIO.parse("acp_train_1912.fasta", "fasta")
lengths1 = [len(s) for s in sequences1]

# Parse FASTA file 2
sequences2 = SeqIO.parse("neg_train_1912.fasta", "fasta")
lengths2 = [len(s) for s in sequences2]

# Plot the lengths
plt.hist([lengths1, lengths2], bins=20, label=["Positive_Length", "Negative_Length"])
plt.title("Cross Validation Dataset",fontsize=8)
plt.xlabel("Sequence Length")
plt.ylabel("Count")
plt.legend()
plt.show()
```
